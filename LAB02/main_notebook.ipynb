Example: suppose you allocated a variable 

```
a = np.zeros(...)
```
So, instead of
```
a = b + c  # will be reallocated, GC needed to free
``` 
You can use: 
```
np.add(b,c,out = a) # puts result in `a`
```

````python
# batch generator
def get_batches(dataset, batch_size):
    X, Y = dataset
    n_samples = X.shape[0]
        
    # Shuffle at the start of epoch
    indices = np.arange(n_samples)
    np.random.shuffle(indices)
    
    for start in range(0, n_samples, batch_size):
        end = min(start + batch_size, n_samples)
        
        batch_idx = indices[start:end]
    
        yield X[batch_idx], Y[batch_idx]
````

````python
def OHE(data):
    n = len(data)
    max_item = np.max(data)
    min_item = np.min(data)
    num_classes = max_item - min_item + 1
    ohe_data = np.zeros((n, num_classes))

    for i, cls in enumerate(range(min_item, max_item + 1)):
        ohe_data[data == cls, i] = 1
    
    assert np.all(np.sum(ohe_data, axis=1) == np.ones(n))
    return ohe_data
````

````python
import time
import copy
````

````python
def pipeline_train(model,
                   criterion,
                   opt_model,
                   opt_config,
                   opt_step,
                   X_train,
                   y_train,
                   X_test,
                   y_test,
                   n_epoch,
                   batch_size = 128):
    
    start = time.time()
    loss_history_train = []
    loss_history_test = []
    acc_history_test = []
    for i in range(n_epoch):
        epoch_loss = []
        model.train()
        for x_batch, y_batch in get_batches((X_train, y_train), batch_size):
            model.zeroGradParameters()

            # Forward
            predictions = model.forward(x_batch)
            loss = criterion.forward(predictions, y_batch)

            # Backward
            dp = criterion.backward(predictions, y_batch)
            model.backward(x_batch, dp)

            # Update weights
            opt_model(model.getParameters(), 
                         model.getGradParameters(), 
                         opt_config,
                         opt_step)      

            epoch_loss.append(loss)
            
        loss_history_train.append(np.mean(epoch_loss))
        epoch_loss = []
        epoch_acc = []
        model.evaluate()
        for x_batch, y_batch in get_batches((X_test, y_test), batch_size):
            predictions = model.forward(x_batch)
            loss = criterion.forward(predictions, y_batch)
            
            labels_pred = np.argmax(predictions, axis=1)
            labels_true = np.argmax(y_batch, axis=1)
            acc = np.mean(labels_pred == labels_true)
            
            epoch_acc.append(acc)
            epoch_loss.append(loss)
            
        loss_history_test.append(np.mean(epoch_loss))  
        acc_history_test.append(np.mean(epoch_acc))
        
    print(f'{model.model_name} computation time: {round(time.time() - start)}')
    return loss_history_train, loss_history_test, acc_history_test
````

````python
def train(model, criterion, opt_model, opt_config, opt_step, X, Y, n_epoch=20, batch_size=1000):
    start = time.time()
    loss_history = []
    
    for i in range(n_epoch):
        epoch_loss = []
        for x_batch, y_batch in get_batches((X, Y), batch_size):

            model.zeroGradParameters()

            # Forward
            predictions = model.forward(x_batch)
            loss = criterion.forward(predictions, y_batch)

            # Backward
            dp = criterion.backward(predictions, y_batch)
            model.backward(x_batch, dp)

            # Update weights
            opt_model(model.getParameters(), 
                         model.getGradParameters(), 
                         opt_config,
                         opt_step)      

            epoch_loss.append(loss)
        loss_history.append(np.mean(epoch_loss))
    print(f'{model.model_name} computation time: {round(time.time() - start)}')
    return loss_history
````

````python
def build_net(name, linear_shapes, activation_module, batch_norm=False):
    net = Sequential()
    net.model_name = name
    net.add(Flatten())
    for (n_in, n_out) in linear_shapes:
        net.add(Linear(n_in, n_out))
        if batch_norm:
            net.add(BatchNormalization())
        net.add(copy.deepcopy(activation_module))
    net.add(LogSoftMax())
    return net
````

````python
def analytics_pipeline(linear_shape, activations, optimizer, optimizer_config, names, title, bn=False, n_epoch=100):
    loss_per_activation = []
    for act_module, model_name in zip(activations, names):
        act_net = build_net(model_name, linear_shapes, act_module, batch_norm=bn)
        loss_train, loss_test, test_acc = pipeline_train(act_net,
                         criterion,
                         optimizer,
                         optimizer_config,
                         {},
                         X_train,
                         y_train_ohe, X_test, y_test_ohe,
                         n_epoch)
        loss_per_activation.append(loss_test)
        
    plt.figure(figsize=(8, 6))
        
    plt.title(title)
    plt.xlabel("#iteration")
    plt.ylabel("loss")
    colors = ['b', 'c', 'r', 'g']
    for act_loss, act_name, color in zip(loss_per_activation, names, colors):
        plt.plot(act_loss, color, label=act_name)
        plt.legend(loc='best')
        plt.yscale('log')
    plt.show()
    return loss_per_activation
````

````python
def plot_comparison(title, loss_1, loss_2, prefix_1, prefix_2):
    plt.figure(figsize=(8, 6))
        
    plt.title(title)
    plt.xlabel("#iteration")
    plt.ylabel("loss")
    colors = ['b', 'c', 'r', 'g']
    names = ['ReLU', 'ELU', 'LeakyReLU', 'SoftPlus']

    for loss, color, name in zip(loss_1, colors, names):
        plt.plot(loss, color, label=f'{prefix_1}_{name}')

    for loss, color, name in zip(loss_2, colors, names):
        plt.plot(loss, color, linestyle='dashed', label=f'{prefix_2}_{name}')

    plt.yscale('log')
    plt.legend(loc='best')
    plt.show()
````
